```python
#mengimport Library yang di butuhkan
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#membaca Dataset
df=pd.read_csv('/content/DATA_PREPARATION.csv')
df.head(5)
```

```python
df_baru = df.copy()
```

```python
# Mencari kolom dengan tipe data non-numerik
kolom_nonnumerik = df_baru.select_dtypes(exclude=['number']).columns
df_nonnumerik = df_baru[kolom_nonnumerik]
df_nonnumerik.head(5)
```

```python
df_fasil = df_baru[['price_in_rp','facilities']]
```

```python
def extract_facilities(df_fasil, col_name):
    facilities_list = df[col_name].str.split(', ').explode().unique().tolist()
    return facilities_list
facilities_list = extract_facilities(df_nonnumerik,'facilities')
```

```python
def count_facilities(df, col_name):
    df['num_facilities'] = df[col_name].str.count(',') + 1
    return df
count_facilities(df_fasil, 'facilities')
```

```python
def encode_facilities(df_fasil, col_name):
    encoded_df = df_fasil[col_name].str.get_dummies(sep=', ')
    return pd.concat([df_fasil, encoded_df], axis=1)
df_fasil = encode_facilities(df_fasil, 'facilities')
```

```python
df_fasil
```

#**ENCODING NONNUMERIK VALUE**

```python
df_nonum = df_nonnumerik[['electricity','property_condition','certificate','furnishing']]
df_nonum.head(10)
```

```python
modus_A = df_nonum['electricity'].mode()[0]
df_nonum.replace('lainnya mah', modus_A, inplace=True)
```

```python
df_nonum['electricity'] = pd.to_numeric(df_nonum['electricity'].str.replace(' mah', ''))
```

```python
df_nonum = pd.get_dummies(df_nonum)
```

```python
df_nonum
```

```python
df_baru = df_baru.drop(['price_in_rp','facilities','electricity','property_condition','certificate','furnishing'], axis=1)
# Menggabungkan dataframe
df_merged = pd.merge(pd.merge(df_baru, df_fasil, left_index=True, right_index=True), df_nonum, left_index=True, right_index=True)
```

```python
df_merged.to_csv('FEATURE_ENGINERING.csv', index= False)
```
